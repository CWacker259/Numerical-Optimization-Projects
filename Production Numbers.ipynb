```python
import pandas as pd 
import matplotlib.pyplot as plt
```

```python
df= pd.read_csv("hamburg plant prod.csv", index_col="SKU") 
df
```

```python
df.sum(axis=0).mean()
```

```python
plt.bar(df.columns,height=df.sum(axis=0),color='orange',edgecolor='black') 
plt.grid(axis='y')
plt.title("Required Production for Hamburg in Years 9 & 10")
plt.ylabel("Hamburg Required Production") 
plt.xlabel("Month")
plt.savefig("Hamburg Production.jpg",dpi=500)
plt.show()
```

```python
mean = pd.Series([df.sum(axis=0).mean()]*8)
```

```python
mean
```

```python
fig, ax1 = plt.subplots(figsize=(12, 8))
bar = df.sum(axis=0).plot(kind='bar', color='orange',edgecolor='black')
line = mean.plot(kind='line', marker='o')
plt.grid(axis='y') 
plt.title("Required Production for Hamburg in Years 9 & 10")
plt.ylabel("Hamburg Required Production") 
plt.xlabel("Month")
plt.legend(labels=["Mean Production","Montly Production"])
plt.savefig("Hamburg Production.jpg",dpi=500)
plt.show()
```
