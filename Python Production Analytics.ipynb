```python
import pandas as pd 
import matplotlib.pyplot as plt 
```

```python
total_df = pd.read_csv("production plan.csv")
index = total_df[['Year','Month']]
```

```python
total_df.index = pd.MultiIndex.from_frame(index)
```

```python
total_df
```

```python
total_df.loc[(2027,"July")]
```

```python
hstore_df = total_df[['H_SKU1_Stor','H_SKU2_Stor','H_SKU3_Stor','H_SKU4_Stor','H_SKU5_Stor']] 
hstore_df = hstore_df.rename(columns={'H_SKU1_Stor':'SKU1','H_SKU2_Stor':'SKU2',
                         'H_SKU3_Stor':"SKU3",'H_SKU4_Stor':'SKU4','H_SKU5_Stor':'SKU5'})
mstore_df = total_df[['M_SKU1_Stor','M_SKU2_Stor','M_SKU3_Stor','M_SKU4_Stor','M_SKU5_Stor']]
mstore_df = mstore_df.rename(columns={'M_SKU1_Stor':'SKU1','M_SKU2_Stor':'SKU2',
                         'M_SKU3_Stor':"SKU3",'M_SKU4_Stor':'SKU4','M_SKU5_Stor':'SKU5'}) 
```

```python
hprod_df = total_df[['H_SKU1_Prod','H_SKU2_Prod','H_SKU3_Prod','H_SKU4_Prod','H_SKU5_Prod']] 
hprod_df = hprod_df.rename(columns={'H_SKU1_Prod':'SKU1','H_SKU2_Prod':'SKU2',
                         'H_SKU3_Prod':"SKU3",'H_SKU4_Prod':'SKU4','H_SKU5_Prod':'SKU5'})
mprod_df = total_df[['M_SKU1_Prod','M_SKU2_Prod','M_SKU3_Prod','M_SKU4_Prod','M_SKU5_Prod']]
mprod_df = mprod_df.rename(columns={'M_SKU1_Prod':'SKU1','M_SKU2_Prod':'SKU2',
                         'M_SKU3_Prod':"SKU3",'M_SKU4_Prod':'SKU4','M_SKU5_Prod':'SKU5'}) 
```

```python
fig, axes = plt.subplots(nrows=2,ncols=1,sharex=True,figsize=(15,10)) 
ax1 = hstore_df.plot(kind='bar',ax=axes[0],title="Hamburg Storage",
                    ylabel="Scooters Quantity",stacked=True) 

ax1.grid(axis='y',linestyle='-',linewidth=0.2)

ax2 = mstore_df.plot(kind='bar',ax=axes[1],title="Madison Storage",
                    xlabel="Year, Month",
                    ylabel="Storage Quantity",stacked=True)
for label in ax2.xaxis.get_ticklabels()[::2]:
    label.set_visible(False)

ax2.grid(axis='y',linestyle='-',linewidth=0.2)  
plt.savefig("Scooter Storage.jpeg",dpi=500)
plt.show() 
```

```python
fig, axes = plt.subplots(nrows=2,ncols=1,sharex=True,figsize=(15,10)) 
ax1 = hprod_df.plot(kind='bar',ax=axes[0],title="Hamburg Production",
                    ylabel="Scooters Required",stacked=True) 

ax1.grid(axis='y',linestyle='-',linewidth=0.2)

ax2 = mprod_df.plot(kind='bar',ax=axes[1],title="Madison Production",
                    xlabel="Year, Month",
                    ylabel="Madison Scooter Production Quantity",stacked=True)
for label in ax2.xaxis.get_ticklabels()[::2]:
    label.set_visible(False)

ax2.grid(axis='y',linestyle='-',linewidth=0.2)  
plt.savefig("Scooter Production.jpeg",dpi=500)
plt.show() 
```
